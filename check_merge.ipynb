```python
import pandas as pd

# Engagement Report 로드
eng = pd.read_excel('What_We_Watched_A_Netflix_Engagement_Report_2023Jan-Jun.xlsx',
                    sheet_name='Engagement', header=5)
eng = eng[['Title','Available Globally?','Release Date','Hours Viewed']].dropna(subset=['Title'])
eng['title_base'] = (eng['Title'].str.lower()
                     .str.replace(r':.*$', '', regex=True)
                     .str.replace(r'\s*//.*$', '', regex=True)
                     .str.strip())

# Netflix 카탈로그 로드
netflix = pd.read_csv('netflix_titles.csv')
netflix['title_clean'] = netflix['title'].str.lower().str.strip()

# 매칭 (제목 기반 left join)
merged = netflix.merge(eng, left_on='title_clean', right_on='title_base', how='inner')
merged_dedup = merged.drop_duplicates(subset='title_clean')

total_netflix = len(netflix)
total_hours   = eng['Hours Viewed'].sum()
matched_hours = merged.groupby('title_clean')['Hours Viewed'].sum().sum()

print(f"Title 매칭률 : {len(merged_dedup):,}건 / {total_netflix:,}건 ({len(merged_dedup)/total_netflix*100:.1f}%)")
```
